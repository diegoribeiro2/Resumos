### 467. Redes Neurais
<h1>Redes Neurais Artificiais (RNA)</h1><h2>O que são?</h2><ul><li>Modelos computacionais inspirados em redes neurais biológicas</li><li>Tentativa de replicar comportamento dos neurônios</li><li>Utilizadas para processamento de texto, visão computacional e reconhecimento de voz</li></ul><h2>Aplicações</h2><ul><li>Identificação de músicas</li><li>Reconhecimento facial</li><li>Classificação de imagens</li><li>Análise de sentimentos em chats</li></ul><h2>Arquitetura</h2><h3>Entrada</h3><ul><li>Dados de entrada</li></ul><h3>Camadas Escondidas</h3><ul><li>Onde ocorre o processamento (Deep Learning)</li></ul><h3>Saída</h3><ul><li>Previsão ou classificação</li></ul><h3>Neurônios</h3><ul><li>Organizados em camadas</li><li>Conectados entre si</li><li>Geram resposta com base nos dados de entrada</li></ul><h2>Próximos passos</h2><ul><li>Explicar com mais detalhes o funcionamento dos neurônios</li></ul>

### 468. Neurônio
<h1>Neurônios e Redes Neurais</h1><h2>Neurônio</h2><ul><li>É como uma regressão múltipla<ul><li>Recebe várias entradas (X)</li><li>Gera uma saída (y)</li></ul></li><li>Cada entrada (X)<ul><li>Tem um peso (W)<ul><li>Como o fator multiplicador na regressão múltipla</li></ul></li><li>É multiplicada pelo seu peso</li></ul></li><li>Tem um viés (bias - b)<ul><li>Funciona como o intercepto na regressão</li></ul></li><li>A saída é:<ul><li>Entrada 1 * Peso 1 + Entrada 2 * Peso 2 + ... + Bias</li></ul></li></ul><h2>Rede Neural Profunda (Deep Learning)</h2><ul><li>Constituída por camadas de neurônios</li><li>Camada inicial<ul><li>Recebe os dados de entrada (X)</li></ul></li><li>Camadas seguintes<ul><li>Recebem a saída dos neurônios da camada anterior</li></ul></li><li>Aprende com experiências/erros anteriores</li><li>Aplicações:<ul><li>Classificação de imagens</li><li>Reconhecimento de objetos</li><li>Processamento de linguagem natural</li></ul></li></ul><h2>Funcionamento</h2><ul><li>Extrai características dos dados</li><li>Entende padrões</li><li>Aprende com os próprios erros</li><li>Aplica o conhecimento para acertar mais</li></ul>

### 469. Função de Ativação
<h1>Funções de ativação em redes neurais</h1><h2>O que é função de ativação</h2><ul><li>Aplicada à soma ponderada feita pelo neurônio</li><li>Determina a saída do neurônio com base na soma ponderada</li></ul><h2>Tipos de funções de ativação</h2><h3>Step Function (Função degrau)</h3><ul><li>Saída binária (0 ou 1)</li><li>Classifica apenas duas categorias</li><li>Problema: não funciona para múltiplas categorias</li></ul><h3>Sigmoid Function</h3><ul><li>Função não linear em formato de S</li><li>Permite saídas contínuas entre 0 e 1</li><li>Pode classificar múltiplas categorias</li><li>Exemplo: classificar imagem como criança (0,9), adulto (0,3) ou idoso (0,1)</li></ul><h2>Para que serve</h2><ul><li>Determinar a saída do neurônio</li><li>Permitir a classificação em duas ou mais categorias</li><li>Em conjunto, os neurônios podem classificar padrões de entrada complexos</li></ul>

### 470. Treinamento
<h1>Redes Neurais - Treinamento</h1><h2>Função custo (erro)</h2><ul><li>MSE (Mean Squared Error)<ul><li>Calcula a diferença entre o valor predito e o valor real</li><li>Quantifica o &quot;erro&quot; do modelo</li></ul></li></ul><h2>Gradiente descendente</h2><ul><li>Atualiza os pesos para minimizar o erro<ul><li>ΔW = Learning rate * Δ * X</li><li>Onde:<ul><li>ΔW = Variação do peso</li><li>Learning rate = Taxa de aprendizado</li><li>Δ = Derivada da função de ativação</li><li>X = Valor da entrada</li></ul></li></ul></li></ul><h2>Objetivo</h2><ul><li>Encontrar pesos ideais que minimizem o erro<ul><li>Modelo se aproxima da realidade</li></ul></li></ul><h2>Outros conceitos</h2><ul><li>Momento<ul><li>Evita mínimos locais</li></ul></li><li>Taxa de aprendizado<ul><li>Define velocidade de aprendizado do modelo</li></ul></li></ul><p>Esse foi um resumo do conteúdo do vídeo focando no conceito de treinamento de redes neurais. Foram abordados os seguintes tópicos principais:</p><ul><li>Função custo (erro)</li><li>Gradiente descendente</li><li>Objetivo do treinamento</li><li>Outros conceitos importantes como momento e taxa de aprendizado</li></ul><p>O mapa mental resume visualmente os principais conceitos, destacando o treinamento como forma de minimizar o erro e fazer o modelo se aproximar da realidade.</p>

<h1>Implementando Redes Neurais em Python</h1><h2>Conteúdo do Dataset</h2><ul><li>Variáveis categóricas:<ul><li>Empréstimo pessoal (variável alvo)</li><li>Conta de título em banco</li><li>Certificado de depósito</li><li>Uso de serviços bancários online</li><li>Uso de cartão de crédito</li></ul></li><li>Variáveis contínuas:<ul><li>Idade</li><li>Experiência</li><li>Renda</li><li>Gastos com cartão de crédito</li><li>Valor da hipoteca</li></ul></li><li>Outras variáveis:<ul><li>Tamanho da família</li><li>Nível educacional</li></ul></li></ul><h2>Pré-processamento dos Dados</h2><ul><li>Excluir variáveis não relevantes (ID, CEP)</li><li>Separar features (X) e target (y)</li><li>Dividir dados em treino e teste</li><li>Padronizar features (standardization)</li></ul><h2>Modelo de Rede Neural</h2><ul><li>Camadas ocultas: 100, 50 e 20 neurônios</li><li>Função de ativação: Sigmoide (logistic)</li><li>Treinar modelo</li><li>Fazer previsões e avaliar métricas:<ul><li>Acurácia: 98%</li><li>F1-score: 89%</li></ul></li></ul><h2>Conclusão</h2><ul><li>Modelo com boa performance</li><li>Desafio: implementar rede neural similar e comparar resultados</li></ul>

```python
import pandas as pd
import numpy as np
```

```python
# Carregando base de dados

df = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx", sheet_name='Data')
```

```python
df.head()
```

```python
# Através da análise das variáveis e do dicionário, algums informações relevantes:


# Variáveis categóricas:
# Personal Loan (Empréstimo Pessoal) - Este cliente aceitou o empréstimo pessoal oferecido na última campanha? Esta é a nossa variável alvo
# Securities Account (Conta de títulos) - O cliente possui conta de títulos no banco?
# CD Account (Conta de CD) - O cliente tem uma conta de certificado de depósito (CD) no banco?
# Online - O cliente usa serviços de banco pela Internet?
# Credit Card - (Cartão de crédito) - O cliente usa um cartão de crédito emitido pelo UniversalBank?

# Variáveis contínuas:

# Age (Idade) Idade do cliente
# Experience (Experiência) - anos de experiência
# Income (Renda) - Renda anual em dólares
# CCAvg (CCAvg) - gasto médio com cartão de crédito
# Mortage (Hipoteca) - Valor da hipoteca da casa

# Variáveis ​​categóricas ordinais:
# Family (Família) - tamanho da família do cliente
# Education (Educação) - nível de educação do cliente


# A variável ID não adiciona nenhuma informação interessante. 
# Não há associação entre o ID de cliente de uma pessoa e o empréstimo, também não fornece nenhuma conclusão geral para futuros clientes de empréstimo em potencial. 
# Podemos desconsiderar essas informações para o modelo.
# Zip code também pode ser desconsiderada, uma vez que a informação de distnância entre zip code pode não ser realidade
```

```python
#Excluindo variáveis não úteis

df = df.drop(columns=["ID"])
```

```python
# Estatística básica das variáveis

df.describe().transpose()
```

```python
# Distruibuição da variáveis target
df["Personal_Loan"].value_counts()
```

```python
# Dividindo preditores (X) e variável target (Y)

X = df.drop(columns=["Personal_Loan"])
Y = df["Personal_Loan"]
```

```python
# Dividindo a base em treino e teste

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=7)

print(f"Shape X_train: {X_train.shape}")
print(f"Shape y_train: {y_train.shape}")
print(f"Shape X_test: {X_test.shape}")
print(f"Shape y_test: {y_test.shape}")
```

```python
# Normalização das variáveis

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Transformanda para dataframe para visualização
X_train = pd.DataFrame(X_train_scaled,columns = X_train.columns)
X_test = pd.DataFrame(X_test_scaled,columns = X_test.columns)
```

```python
# Classificador Multi Layer Perceptron
# Hidder_layer sizes camadas escondidas
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

clf = MLPClassifier( hidden_layer_sizes=(100, 50, 20),activation='logistic')
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
print(f"Acurácia: {accuracy_score(y_test, y_pred)*100}%")
```

```python
print(f"F1 Score: {f1_score(y_test, y_pred)*100}%")
```

```python
import pandas as pd
```

```python
df = pd.read_excel("churn.xlsx")
```

```python
df.head()
```

```python
#Excluindo variáveis não úteis

df = df.drop(columns=["customerID"])
```

```python
df = df.dropna()
```

```python
df.dtypes
```

```python
df.isnull().any()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"],errors='coerce')
```

```python
cat_vars = ['PhoneService',"Contract","PaperlessBilling","PaymentMethod"]
num_vars = ["tenure","MonthlyCharges","TotalCharges"]
modeling_vars = cat_vars + num_vars
```

```python
X = df.filter(modeling_vars).copy()
y = df['Churn'].copy()
y = pd.DataFrame(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(y.Churn) 
y.Churn = le.transform(y.Churn)
```

```python
X = pd.get_dummies(X, dtype=int)
```

```python
# Distruibuição da variáveis target
df["Churn"].value_counts()
```

```python
# Dividindo a base em treino e teste

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

print(f"Shape X_train: {X_train.shape}")
print(f"Shape y_train: {y_train.shape}")
print(f"Shape X_test: {X_test.shape}")
print(f"Shape y_test: {y_test.shape}")
```

```python
# Normalização das variáveis

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Transformanda para dataframe para visualização
X_train = pd.DataFrame(X_train_scaled,columns = X_train.columns)
X_test = pd.DataFrame(X_test_scaled,columns = X_test.columns)
```

```python
# Classificador Multi Layer Perceptron

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score

clf = MLPClassifier(hidden_layer_sizes=(100, 50, 20),activation='logistic')
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
print(f"Acurácia: {accuracy_score(y_test, y_pred)*100}%")
```

```python
y_pred = clf.predict(X_test)
print(f"Acurácia: {f1_score(y_test, y_pred)*100}%")
```
