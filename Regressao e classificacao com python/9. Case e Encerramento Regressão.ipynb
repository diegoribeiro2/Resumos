```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
seguro=pd.read_csv("insurance.csv")
seguro.head()
```

```python
seguro.info()
```

```python
seguro.dropna(inplace=True)
```

```python
# Yes e No
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

le.fit(seguro.sexo)
seguro.sexo= le.transform(seguro.sexo)

le.fit(seguro.fumante)
seguro.fumante= le.transform(seguro.fumante)

le.fit(seguro.regiao)
seguro.regiao = le.transform(seguro.regiao)
```

```python
seguro = pd.DataFrame(seguro)
```

```python
seguro.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

seguro_norm= pd.DataFrame(scaler.fit_transform(seguro), index=seguro.index, columns=seguro.columns)
seguro_norm.head()
```

```python
import statsmodels.formula.api as smf
function="custos_seguro~idade+sexo+imc+quantidade_filhos+fumante+regiao"
model = smf.ols(formula=function, data=seguro_norm).fit()
print(model.summary())
```

```python
import statsmodels.formula.api as smf
function="custos_seguro~idade+imc+quantidade_filhos+fumante+regiao"
model = smf.ols(formula=function, data=seguro_norm).fit()
print(model.summary())
```

```python
x = seguro[['idade','imc','quantidade_filhos','fumante','regiao']]
y= seguro_norm['custos_seguro']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)

lr=LinearRegression()
lr.fit(x_train,y_train)
LinearRegression()
```

```python
from sklearn import metrics
r_sq = lr.score(x,y)
print(r_sq)
```

```python
y_pred_train = lr.predict(x_train)
print("MAE:", metrics.mean_absolute_error(y_train, y_pred_train))
```

```python
y_pred_test = lr.predict(x_test)
print("MAE:", metrics.mean_absolute_error(y_test, y_pred_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
from sklearn import metrics
r_sq = rf.score(x,y)
print(r_sq)
y_pred_train = rf.predict(x_train)
print("MAE", metrics.mean_absolute_error(y_train, y_pred_train))

y_pred_test = rf.predict(x_test)
print("MAE", metrics.mean_absolute_error(y_test, y_pred_test))
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()
ada.fit(x_train,y_train)
```

```python
r_sq = ada.score(x,y)
print(r_sq)
y_pred_train = ada.predict(x_train)
print("MAE", metrics.mean_absolute_error(y_train, y_pred_train))

y_pred_test = ada.predict(x_test)
print("MAE", metrics.mean_absolute_error(y_test, y_pred_test))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
grb = GradientBoostingRegressor()
grb.fit(x_train,y_train)
```

```python
r_sq = grb.score(x,y)
print(r_sq)
y_pred_train = grb.predict(x_train)
print("MAE", metrics.mean_absolute_error(y_train, y_pred_train))

y_pred_test = grb.predict(x_test)
print("MAE", metrics.mean_absolute_error(y_test, y_pred_test))
```

```python
# Tunning
from sklearn.model_selection import GridSearchCV

parameters = {
               'max_depth':[5],
               'min_samples_leaf':[4],
               'min_samples_split':[2],
               'n_estimators':[100]}

grid_search = GridSearchCV(grb, parameters,scoring="r2",cv=2, n_jobs=-1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print(grid_search.best_estimator_)
print(grid_search.best_params_)
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model.get_params()
```

```python
grb_tunned = GradientBoostingRegressor(alpha = 0.9,
ccp_alpha = 0.0,
criterion = 'friedman_mse',
init = None,
learning_rate= 0.1,
loss = 'squared_error',
max_depth= 5,
max_features= None,
max_leaf_nodes= None,
min_impurity_decrease= 0.0,
min_samples_leaf= 4,
min_samples_split= 2,
min_weight_fraction_leaf= 0.0,
n_estimators= 100,
n_iter_no_change= None,
random_state= None,
subsample= 1.0,
tol= 0.0001,
validation_fraction= 0.1,
verbose= 0,
warm_start= False)
                                    
```

```python
grb_tunned.fit(x_train,y_train)

```

```python
r_sq = grb_tunned.score(x,y)
r_sq

```

```python
y_pred_train= grb_tunned.predict(x_train)
print("MAE:", metrics.mean_absolute_error(y_train,y_pred_train))

x_pred_test=grb_tunned.predict(x_test)
print("MAE:", metrics.mean_absolute_error(y_test,y_pred_test))
```
