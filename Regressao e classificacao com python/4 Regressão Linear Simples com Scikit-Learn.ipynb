```python
import pandas as pd
import numpy as np
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
imoveis = pd.read_excel('imoveis.xlsx')
imoveis.head()
```

```python
imoveis.describe()
```

```python
imoveis.info()
```

```python
# Linha de normalidade kde
sns.histplot(imoveis,x ="area", kde=True)
```

```python
sns.histplot(imoveis,x ="preco", kde=True)
```

```python
# Análises Bi-variáveis
imoveis.corr()
```

```python
# Reta da regressão
sns.regplot(x="area", y="preco", data=imoveis)
```

```python
imoveis.describe()
```

```python
# Medidas de normalização, aplicando uma transformação na escala de 0 até 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

imoveis_norm= pd.DataFrame(scaler.fit_transform(imoveis), index = imoveis.index, columns=imoveis.columns)
```

```python
imoveis_norm.head()
```

```python
imoveis_norm.describe()
```

```python
sns.regplot(x="area", y="preco", data=imoveis_norm)
```

```python
import statsmodels.formula.api as smf
```

```python
function="preco~area"
model = smf.ols(formula=function, data=imoveis_norm).fit()
print(model.summary())
```

```python
x= imoveis_norm[["area"]]
                
y= imoveis_norm[["preco"]]
```

```python
# Montando o modelo de MachineLearn
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.20, random_state=42)

lr= LinearRegression()

lr.fit(x_treino, y_treino)

LinearRegression()
```

```python
# Métricas de sucesso do meu treino
r_sq = lr.score(x_treino, y_treino)
print('Coeficiente de Determinação (R²):', r_sq)
```

```python
# Algoritmo de regressão linear vai treinar o meu x treino 
y_pred_treino = lr.predict(x_treino)

# Quanto meu treino pred está distante do meu y  de treino original
print("MAE:", metrics.mean_absolute_error(y_treino, y_pred_treino))
# Erro métrico quadratico 
print("MSE:", metrics.mean_squared_error(y_treino, y_pred_treino))
# Raiz quadrada do erro metrico quadrado
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_treino, y_pred_treino)))
```

```python
lr.fit(x_teste, y_teste)
```

```python
y_pred_teste = lr.predict(x_teste)
```

```python
print("MAE:", metrics.mean_absolute_error(y_teste, y_pred_teste))
print("MSE:", metrics.mean_squared_error(y_teste, y_pred_teste))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_teste, y_pred_teste)))
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
alunos = pd.read_excel("aprovacao_alunos2.xlsx")
alunos.head()
```

```python
alunos.describe()
```

```python
alunos.corr()
```

```python
alunos.info()
```

```python
sns.histplot(data=alunos, x="ano_2015", kde=True)
```

```python
sns.histplot(data=alunos, x="ano_2016", kde=True)
```

```python
sns.regplot(x="ano_2015", y="ano_2016",data=alunos)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

alunos_norm = pd.DataFrame(scaler.fit_transform(alunos), index=alunos.index, columns=alunos.columns)
```

```python
alunos_norm.head()
```

```python
alunos_norm.describe()
```

```python
import statsmodels.formula.api as smf
```

```python
function="ano_2016~ano_2015-1"
model = smf.ols(formula=function, data=alunos_norm).fit()
print(model.summary())
```

```python
x= alunos_norm[["ano_2015"]]

y=alunos_norm[["ano_2016"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

lr= LinearRegression()

lr.fit(x_train,y_train)
```

```python
r_sq=lr.score(x_train,y_train)
print(r_sq)
```

```python
y_pred_train = lr.predict(x_train)
print("MAE:", metrics.mean_absolute_error(y_train, y_pred_train))
print("MSE:", metrics.mean_squared_error(y_train, y_pred_train))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_train, y_pred_train)))
```

```python
y_pred_test = lr.predict(x_test)
print("MAE:", metrics.mean_absolute_error(y_test, y_pred_test))
print("MSE:", metrics.mean_squared_error(y_test, y_pred_test))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
```
