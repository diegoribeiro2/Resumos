### 279 Filtrando os Dados I
<h1>Filtros em Python</h1><h2>Introdução</h2><ul><li>Objetivo: Aprender sobre filtros no Python para análise de dados</li><li>Serão mostradas formas de filtrar no Excel, SQL e Python para comparação</li><li>A lógica é a mesma, o que muda é a sintaxe</li></ul><h2>Premissas</h2><ul><li>Budget para aluguel: até 4 mil euros</li><li>Precisa aceitar animais de estimação</li><li>Apartamento com 2 ou 3 quartos</li></ul><h2>Filtros no Excel</h2><ul><li>Número -&gt; Menor ou igual a 4 mil</li><li>Texto -&gt; É igual a &quot;Aceita&quot;</li><li>Número -&gt; Entre 2 e 3 (quartos)</li></ul><h2>Filtros no SQL</h2><ul><li>WHERE total &lt;= 4000</li><li>AND animal = &quot;Aceita&quot;</li><li>AND (rooms BETWEEN 2 AND 3)</li></ul><h2>Filtros no Python</h2><ul><li>aluguel_casa[total] &lt;= 4000</li><li>&amp; (aluguel_casa[&quot;animal&quot;] == &quot;Aceita&quot;)</li><li>| (aluguel_casa[&quot;rooms&quot;] == 2)</li><li>| (aluguel_casa[&quot;rooms&quot;] == 3)</li></ul><h2>Comparação</h2><ul><li>Mesma lógica nos 3 casos</li><li>Sintaxe muda entre as ferramentas</li><li>Python mais flexível para customização</li></ul><h2>Resultados</h2><ul><li>Linhas reduzidas de 6.080 para 1.407 após filtros</li><li>Premissas respeitadas:<ul><li>Valor máximo 4 mil</li><li>Máximo de 3 quartos</li></ul></li><li>Dataset reduzido para casos que atendem necessidades</li></ul>


```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept")]
aluguel_casa.head()
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept") & ((aluguel_casa["rooms"]==2)|(aluguel_casa["rooms"]==3))]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

### 280 Filtrando os Dados II

<h1>Filtros em Python</h1><h2>Tema central</h2><p>Outra forma de fazer filtros em Python de maneira simples, semelhante ao SQL, utilizando a função .query</p><h2>Funcionalidades do .query</h2><ul><li>Sintaxe simples, utiliza menos palavras que os filtros originais</li><li>Permite fazer filtros parecidos com o SQL</li><li>Basta passar as condições entre parênteses após o .query</li></ul><h3>Exemplos</h3><pre><code>aluguel_casa.query(&quot;total &lt;= 4000 &amp; aceita_animais == True&quot;)</code></pre><ul><li>Utiliza os operadores lógicos do Python (&amp; para E)</li><li>Aspas simples para strings</li></ul><pre><code>aluguel_casa.query(&quot;total &lt;= 4000 &amp; (quartos == 2 | quartos == 3)&quot;)  </code></pre><ul><li>Operador OR representado por |</li><li>Parenteses para agrupar condições</li></ul><h2>Vantagens</h2><ul><li>Código mais limpo e menos verboso</li><li>Economiza tempo e linhas de código</li><li>Fácil utilização para quem já conhece SQL</li></ul><h2>Conclusão</h2><p>O .query é uma função que simplifica a criação de filtros em Python, aproximando da sintaxe SQL, sendo menos verbosa. Recomendada para substituir filtros complexos.</p>


```python
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3")
aluguel_casa.head()
```

### 281 Agrupando os dados

<h1>Funções de agregação em análise de dados</h1><h2>Tema central</h2><p>Uso de funções de agregação (como média, soma etc) para analisar e resumir dados em ferramentas diferentes:</p><h3>Excel</h3><ul><li>Uso de tabelas dinâmicas<ul><li>Agrupar por colunas (cidade, quartos, banheiros)</li><li>Calcular médias de outra coluna (preço)</li></ul></li></ul><h3>SQL</h3><ul><li>Função <code>AVG()</code> para média</li><li>Cláusula <code>GROUP BY</code> para agrupar</li></ul><h3>Python</h3><ul><li>Método <code>groupby()</code> para agrupar</li><li>Funções como <code>mean()</code> para agregação</li><li>Mesma lógica do Excel e SQL</li></ul><h2>Comparações</h2><ul><li>Lógica é a mesma nas 3 ferramentas</li><li>Sintaxe muda entre as ferramentas</li><li>Erros são comuns ao aprender</li></ul><h2>Observações</h2><ul><li>Exemplos práticos em vídeo para dados de aluguel de casas</li><li>Filtros aplicados antes para focar análise</li><li>Agrupamento por múltiplas colunas também visto</li></ul><p>Esse resumo traz o tema central, os tópicos abordados e compara as 3 ferramentas analisadas, com observações adicionais. Formatei com Markdown para facilitar a visualização.</p>


```python
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
agg_bath_room_price = aluguel_casa.groupby(["bathroom","rooms"])["total"].agg("mean").reset_index()
agg_bath_room_price
```
