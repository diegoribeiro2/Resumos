###  373 - Categorical Enconding - Teórico
<h1>Tratamento de variáveis categóricas (tema central)</h1><h2>Variáveis numéricas</h2><ul><li>Modelos conseguem operar nativamente</li><li>Mas é preciso:<ul><li>Padronização (ex: z-score, min-max scale)</li><li>Tratamento de outliers</li></ul></li></ul><h2>Variáveis categóricas</h2><ul><li>Modelos não entendem texto</li><li>Opções de tratamento:<ul><li><strong>One-hot encoding</strong>:<ul><li>Criar coluna para cada categoria</li><li>Preencher com 1/0 conforme presença da categoria</li></ul></li><li><strong>Ordinal encoding</strong>:<ul><li>Atribuir números para cada categoria (ex: baixo = 1, médio = 2, alto = 3)</li></ul></li></ul></li></ul><p>O vídeo introduz as duas principais formas de tratar variáveis categóricas para uso em modelos de machine learning: one-hot encoding e ordinal encoding.</p><p>Primeiro, faz um disclaimer sobre variáveis numéricas, explicando que os modelos conseguem operar nativamente, mas beneficiam-se de padronização e tratamento de outliers.</p><p>Depois, explica que variáveis categóricas precisam ser tratadas pois os modelos não entendem texto. Apresenta então as duas opções de encoding.</p><p>One-hot encoding envolve criar colunas binárias para cada categoria, marcando com 1/0 se ela está presente na amostra. Já ordinal encoding substitui as categorias por números ordenados.</p><p>O vídeo termina dizendo que os detalhes serão mostrados na prática no próximo vídeo.</p>

###  375 - Categorical Enconding - Labelenconding
<h1>Tratamento de variáveis categóricas</h1><h2>Label Encoder</h2><ul><li>Substitui texto por valor numérico</li><li>Funciona para alguns modelos mas não para outros (ex: regressão)</li><li>Usa menos memória que one-hot encoding</li></ul><h2>Em Pandas</h2><ul><li>Converte para tipo &quot;category&quot;</li><li>Atribui códigos automaticamente (0, 1, 2...)</li><li>Permite mapear códigos para categorias</li></ul><h2>Em Scikit-Learn</h2><ul><li>OriginalEncoder</li><li>segue padrão fit() e transform()</li><li>mapeia categorias para números durante fit</li><li>transforma dados de acordo durante transform()</li><li>handle_unknown permite lidar com novas categorias</li></ul><h2>Próximos tópicos</h2><ul><li>Diferentes formas de fazer filtros em Python</li></ul>

```python
# Usando pandas
# Modelos preditivos Sklearn, Modelos não supervisionado Pandas
# Convertendo para categoria
df_cat = df.copy()

df_cat[categorical_columns] = df_cat[categorical_columns].astype('category')
df_cat.info()
```

```python
{col:dict(enumerate(df_cat['smoker'].cat.categories))for col in categorical_columns}
```

```python
df_cat['smoker'].cat.codes
```

```python
# Modificando colunas
df_cat[categorical_columns]=df_cat[categorical_columns].apply(lambda x: x.cat.codes)
```

```python
df_cat.head()
```

```python
# Usando factorize
df['sex'].factorize()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Instanciando OriginalEncoder
encoder = OrdinalEncoder().fit(df[categorical_columns])
```

```python
# Mostrando atributos
encoder.categories_
```

```python
# Transformando dados esperados
df_cat[categorical_columns]= encoder.transform(df[categorical_columns])
```
